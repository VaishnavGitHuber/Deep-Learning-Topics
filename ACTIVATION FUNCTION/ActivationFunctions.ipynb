### What is a Neural Network Activation Function?
* An Activation Function decides whether a neuron should be activated or not. This means that it will decide whether the neuron's input to the network is important or not in the process of prediction using simpler mathematical operations.
* Most popular Activation Functions
1. Sigmoid Activation Function
2. Tanh Activation function
3. Relu Activation Function
4. Leaky Relu Activation Function
* Our Aim of today is to understand and Visualise the Activation functions in a better way by using visuallisation tenchniques.

```python
import matplotlib.pyplot as plt
import math
```

### Sigmoid Activation function
* Values rangesd from 0 - 1
* It is always used in output layer if it is binary classfication. and Softmax if it is multiclass classification
* Works well for binary classification
* Faces problem due to the Vanishing Gradient :  Vanishing gradient problem is a phenomenon that occurs during the training of deep neural networks, where the gradients that are used to update the network become extremely small or "vanish" as they are backpropogated from the output layers to the earlier layers.

```python
# Sigmoid activation function
def sigmoid(x):
  return 1 / (1+math.exp(-x))
```

```python
# I am taking some sigmoid function values and goind to plot
x = list(range(-10,10+1,1))
y = list(map(sigmoid,x))
```

```python
# Visualisation of the Sigmoid function
plt.title("Sigmoid Function")
plt.plot(x,y,marker="*")
plt.grid(visible=True)
plt.show()
```

### Tanh Activation function
* Ranges from -1 to 1
* This activation function faces Vanishing gradient

```python
# tanh function
def tanh(x):
  return (math.exp(x) - math.exp(-x))/(math.exp(x) + math.exp(-x))

# taking 2 variables for visualisation
x = range(-20,20+1,1)
y = list(map(tanh,x))
```

```python
# plotting the tanh function
plt.title("Tanh function")
plt.plot(x,y,marker="o",markersize=3)
plt.grid(visible=True)
plt.show()
```

### Relu Activation function(Rectified Linear unit)
* Computationally very efficent
* Commonly used in Hidden layer
* Faces: Dead neuron problem

```python
# Relu activation function
def relu(x):
  return max(0,x)

# Taking some example for visualisation
x = range(-10,10+1,1)
y = list(map(relu,x))
```

```python
# Visualisation of the relu activation function
plt.title("Relu activation function")
plt.plot(x,y,marker="o",markersize=4)
plt.grid(visible=True)
plt.show()
```

### Leaky Relu Activation Function

```python
# Leaky Relu Activation Function
def leakyRelu(x):
  return max(0.1*x,x)

# Simple example for visualisation
x = range(-10,10,1)
y = list(map(leakyRelu,x))
```

```python
plt.title("Leaky Relu Activation function")
plt.plot(x,y)
plt.grid(visible=True)
plt.show()
```
